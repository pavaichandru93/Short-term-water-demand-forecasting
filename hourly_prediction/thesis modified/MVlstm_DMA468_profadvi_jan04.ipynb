```python
data = pd.read_csv("/content/gdrive/My Drive/Thesis/dataframe_flow_variable_60min.csv")
data['Unnamed: 0'] = pd.to_datetime(data['Unnamed: 0'])
data.rename( columns={'Unnamed: 0':'Date_time','468':'DMA_468'}, inplace=True )
```

```python
data_DMA468 = data[['Date_time','DMA_468','tempC','HeatIndexC','FeelsLikeC','WindChillC','windspeedKmph','sunHour','precipMM','humidity']]
data_DMA468.shape
```

```python
data_DMA468_indexed = data_DMA468.set_index('Date_time')
data_DMA468_indexed[['DMA_468','tempC','HeatIndexC','FeelsLikeC','WindChillC','windspeedKmph','sunHour','precipMM','humidity']].plot(subplots=True)
plt.figure(figsize=(20, 20))
```

```python
# Making hourly data 
hourly_data = data_DMA468_indexed.resample('h').sum()

#Train and test split
#split = 442
train = hourly_data[:-442]
test = hourly_data[-442:]
```

```python
#Separate dates for future plotting
trainset = train.reset_index()
train_dates = pd.to_datetime(trainset['Date_time'])
len(train_dates)
```

```python
#Variables for training
cols = list(train)[:10]
df_for_training = train[cols]

#LSTM uses sigmoid and tanh that are sensitive to magnitude so values need to be normalized
# normalize the dataset
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
df_for_training_scaled = scaler.transform(df_for_training)
df_for_training_scaled.shape
```

```python
#Empty lists to be populated using formatted training data
trainX = []
trainY = []

n_future = 1   # Number of days we want to look into the future based on the past days.
n_past = 24  # Number of past days we want to use to predict the future.

#Reformat input data into a shape: (n_samples x timesteps x n_features)
#In my example, my df_for_training_scaled has a shape (12823, 5)
#12823 refers to the number of data points and 5 refers to the columns (multi-variables).
for i in range(n_past, len(df_for_training_scaled) - n_future +1):
    trainX.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
    trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

trainX, trainY = np.array(trainX), np.array(trainY)

print('trainX shape == {}.'.format(trainX.shape))
print('trainY shape == {}.'.format(trainY.shape))

trainX, trainY = np.array(trainX), np.array(trainY)
```

```python
model = Sequential()
model.add(LSTM(60, activation='relu',input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LeakyReLU(alpha=0.9)) 
model.add(Dropout(0.17))
model.add(LSTM(15,return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(trainY.shape[1]))

model.compile(optimizer='adam', loss='mae')
model.summary()


# fit the model
history = model.fit(trainX, trainY, epochs=100, batch_size=10000, validation_split=0.2, verbose=1)
```

```python
# Plot train loss and validation loss
def plot_loss (history, model_name):
    plt.figure(figsize = (10, 6))
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Train vs Validation Loss for ' + model_name)
    plt.ylabel('Loss')
    plt.xlabel('epoch')
    plt.legend(['Train loss', 'Validation loss'], loc='upper right')
```

```python
plot_loss(history,'LSTM')
```

```python
n_past = 24
n_days_for_prediction=264  #let us predict past 15 days

predict_period_dates = pd.date_range(list(train_dates)[-n_past], periods=n_days_for_prediction, freq='1H').tolist()

#Make prediction
prediction = model.predict(trainX[-n_days_for_prediction:])
prediction_copies = np.repeat(prediction, df_for_training.shape[1], axis=-1)
y_pred_future = scaler.inverse_transform(prediction_copies)[:,0]
```

```python
# Convert timestamp to date
forecast_dates = []
for time_i in predict_period_dates:
    forecast_dates.append(time_i)
    
df_forecast = pd.DataFrame({'Date_time':np.array(forecast_dates), 'DMA_468_pred':y_pred_future})
df_forecast['Date_time']=pd.to_datetime(df_forecast['Date_time'])

# making predicted df

df1=df_forecast[24:]
df1= df1.reset_index()
df1= df1.drop(['index'], axis=1)

# test dataset

t1 = test.reset_index()
t1 = t1[['Date_time','DMA_468']]

# final df
fd=pd.merge(t1, df1, on="Date_time")
```
